Question 2: Healthcare Analytics Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Step 2: Create the Dataset

```python
data = {
    "Patient_ID": ["P01","P02","P03","P04","P05","P06","P07","P08","P09","P10",
                   "P11","P12","P13","P14","P15","P16","P17","P18","P19","P20"],

    "Age": [45,32,55,28,60,40,52,35,48,30,
            62,38,50,27,58,42,46,34,53,29],

    "Sugar_Level": [145,98,180,90,200,120,165,105,155,95,
                    210,115,170,88,190,130,150,100,175,92],

    "BP": [130,118,145,110,150,125,140,120,135,115,
           155,122,142,108,148,128,132,116,144,112],

    "BMI": [28,23,31,22,34,26,30,24,29,23,
            35,25,30,21,33,27,28,24,31,22],

    "Disease_Status": ["Yes","No","Yes","No","Yes","No","Yes","No","Yes","No",
                       "Yes","No","Yes","No","Yes","No","Yes","No","Yes","No"]
}

df = pd.DataFrame(data)
```

Step 3: Display the Dataset

```python
print(df)
```

Step 4: Display First and Last 5 Rows

```python
print("First 5 Rows")
print(df.head())

print("\nLast 5 Rows")
print(df.tail())
```

Step 5: Rows and Columns

```python
print("Shape:", df.shape)
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

Step 6: Dataset Information

```python
df.info()

print("\nData Types")
print(df.dtypes)
```

Step 7: Missing and Duplicate Values

```python
print("Missing Values")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())
```

Step 8: Statistical Summary

```python
print(df.describe())

print("\nMean")
print(df.mean(numeric_only=True))

print("\nMedian")
print(df.median(numeric_only=True))

print("\nMinimum")
print(df.min(numeric_only=True))

print("\nMaximum")
print(df.max(numeric_only=True))

print("\nStandard Deviation")
print(df.std(numeric_only=True))
```

Step 9: Data Visualizations

```python
# Bar Chart
plt.figure(figsize=(8,5))
plt.bar(df["Patient_ID"], df["Sugar_Level"])
plt.title("Patient Sugar Levels")
plt.xlabel("Patient ID")
plt.ylabel("Sugar Level")
plt.show()

# Histogram
plt.figure(figsize=(6,4))
plt.hist(df["BMI"], bins=5)
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()

# Scatter Plot
plt.figure(figsize=(6,5))
plt.scatter(df["Age"], df["Sugar_Level"])
plt.title("Age vs Sugar Level")
plt.xlabel("Age")
plt.ylabel("Sugar Level")
plt.show()

# Box Plot
plt.figure(figsize=(5,5))
plt.boxplot(df["BP"])
plt.title("Blood Pressure Box Plot")
plt.ylabel("Blood Pressure")
plt.show()
```

Step 10: Pattern Identification

```python
print(df.groupby("Disease_Status")[["Age","Sugar_Level","BP","BMI"]].mean())
```

Step 11: Observations
1. Patients with disease generally have higher sugar levels.
2. Higher blood pressure is mostly associated with disease.
3. Patients with higher BMI are more likely to have disease.
4. Older patients tend to have higher health risk.
5. No missing values or duplicate records were found.

Step 12: Conclusion
The EDA shows that Age, Sugar Level, Blood Pressure, and BMI are important factors influencing disease status. Patients with higher values in these health indicators are more likely to have the disease. The dataset is clean and suitable for analysis.
