Step 32: Data Visualizations

```python
# Bar Chart
plt.figure(figsize=(8,5))
plt.bar(df["Student_ID"], df["Attendance"])
plt.title("Student Attendance")
plt.xlabel("Student ID")
plt.ylabel("Attendance")
plt.show()

# Histogram
plt.figure(figsize=(6,4))
plt.hist(df["Study_Hours"], bins=6)
plt.title("Study Hours Distribution")
plt.xlabel("Study Hours")
plt.ylabel("Frequency")
plt.show()

# Scatter Plot
plt.figure(figsize=(6,5))
plt.scatter(df["Study_Hours"], df["Internal_Marks"])
plt.title("Study Hours vs Internal Marks")
plt.xlabel("Study Hours")
plt.ylabel("Internal Marks")
plt.show()

# Box Plot
plt.figure(figsize=(5,5))
plt.boxplot(df["Attendance"])
plt.title("Attendance Box Plot")
plt.ylabel("Attendance")
plt.show()
```

Step 33: Identify Important Patterns

```python
print(df.groupby("Result")[["Attendance", "Study_Hours", "Internal_Marks"]].mean())
```
